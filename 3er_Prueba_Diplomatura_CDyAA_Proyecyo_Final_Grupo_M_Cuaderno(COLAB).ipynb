```python
from scipy import stats
import scikit_posthocs as sp

# Perform Kruskal-Wallis test
# We need to group the target variable by 'censo_anio'
groups = df_filtered.groupby('censo_anio')['participo_alteracion_orden_ult_anio_id'].apply(list)

# Perform Kruskal-Wallis test
kw_statistic, kw_p_value = stats.kruskal(*groups)

print(f"Kruskal-Wallis Test:")
print(f"  Statistic: {kw_statistic:.4f}")
print(f"  P-value: {kw_p_value:.4f}")

# Perform Dunn's post-hoc test if Kruskal-Wallis test is significant
if kw_p_value < 0.05:
    print("\nKruskal-Wallis test is significant, performing Dunn's post-hoc test:")
    # Create a list of arrays, one for each group
    data = [np.array(group) for group in groups]
    # Perform Dunn's test
    dunn_results = sp.posthoc_dunn(data, p_adjust='bonferroni')
    display(dunn_results)
else:
    print("\nKruskal-Wallis test is not significant, no post-hoc test needed.")
```

```python
# LIBRERIAS

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings('ignore')

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

# PARAMETROS

rnd_state = 42
```

## Arboles de Desición: LGBM y CatBoost

## LGBMClassifier

#### Temporal Holdout con Rolling window

Rolling window 2019–2021 para validar estabilidad.

GridSearchCV en 2020–2021 para tuning de hiperparámetros (num_leaves, learning_rate, max_depth, min_child_samples).

Entrenamiento final (2002–2021) y test en 2022 con mejores hiperparámetros


Métricas: Balanced Accuracy, F1 Macro, PR AUC, ROC AUC.

Curvas PR por clase en el test 2022.

Gráfico de evolución temporal de todas las métricas.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import (
    classification_report, confusion_matrix,
    f1_score, balanced_accuracy_score,
    precision_recall_curve, auc,
    average_precision_score, roc_auc_score
)
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV
from lightgbm import early_stopping, log_evaluation # Import necessary callbacks

# -----------------------------
# 1. Separar features y target
# -----------------------------
X = df_modified.drop(['participo_alteracion_orden_ult_anio_id'], axis=1, errors='ignore')
y = df_modified['participo_alteracion_orden_ult_anio_id']

# Identificar columnas categóricas
categorical_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
for col in categorical_cols:
    X[col] = X[col].astype('category')  # LightGBM puede usar categóricas nativas

# -----------------------------
# 2. Rolling Window Validation
# -----------------------------
years = sorted(X['censo_anio'].unique())
print("Años disponibles:", years)

metrics_per_year = []

for val_year in range(2019, 2022):  # validamos de 2019 a 2021
    print(f"\n>>> Entrenando hasta {val_year-1}, validando en {val_year}")

    # Train = todo hasta año previo
    X_train = X[X['censo_anio'] < val_year].drop('censo_anio', axis=1)
    y_train = y[X_train.index]

    # Validation = año actual
    X_val = X[X['censo_anio'] == val_year].drop('censo_anio', axis=1)
    y_val = y[X_val.index]

    # Modelo base
    lgbm_model = LGBMClassifier(
        n_estimators=500,
        learning_rate=0.05,
        num_leaves=64,
        class_weight="balanced",
        objective="multiclass",
        random_state=42
    )

    # Define callbacks for early stopping and logging
    callbacks = [early_stopping(stopping_rounds=30, verbose=-1), log_evaluation(period=100)]

    lgbm_model.fit(
        X_train, y_train,
        categorical_feature=[c for c in categorical_cols if c in X_train.columns],
        eval_set=[(X_val, y_val)],
        callbacks=callbacks # Use callbacks parameter
    )

    # Predicciones
    y_val_pred = lgbm_model.predict(X_val)
    y_val_proba = lgbm_model.predict_proba(X_val)

    # Métricas
    bal_acc = balanced_accuracy_score(y_val, y_val_pred)
    f1 = f1_score(y_val, y_val_pred, average="macro")

    # PR AUC y ROC AUC
    classes = lgbm_model.classes_
    y_val_bin = label_binarize(y_val, classes=classes)
    pr_auc = average_precision_score(y_val_bin, y_val_proba, average="macro")
    roc_auc = roc_auc_score(y_val_bin, y_val_proba, average="macro")

    print(f"Balanced Accuracy: {bal_acc:.4f} | F1 Macro: {f1:.4f} | PR AUC: {pr_auc:.4f} | ROC AUC: {roc_auc:.4f}")

    metrics_per_year.append({
        "validation_year": val_year,
        "balanced_accuracy": bal_acc,
        "f1_macro": f1,
        "pr_auc": pr_auc,
        "roc_auc": roc_auc
    })

# Convertir a DataFrame
metrics_df = pd.DataFrame(metrics_per_year)
print("\nResumen Rolling Window CV:")
print(metrics_df)

# -----------------------------
# 3. GridSearchCV en 2020–2021 (para tuning)
# -----------------------------
print("\n>>> Ajustando hiperparámetros con GridSearchCV (2020–2021)")

X_tune = X[X['censo_anio'].isin([2020, 2021])].drop('censo_anio', axis=1)
y_tune = y[X_tune.index]

# Identify categorical columns in X_tune
categorical_cols_tune = X_tune.select_dtypes(include=['object', 'category']).columns.tolist()

param_grid = {
    "num_leaves": [31, 64, 128],
    "learning_rate": [0.01, 0.05, 0.1],
    "max_depth": [-1, 10, 20],
    "min_child_samples": [20, 50, 100]
}

lgbm_grid = LGBMClassifier(
    n_estimators=500,
    class_weight="balanced",
    objective="multiclass",
    random_state=42
)

grid = GridSearchCV(
    estimator=lgbm_grid,
    param_grid=param_grid,
    scoring="f1_macro",
    cv=3,
    n_jobs=-1
)

grid.fit(
    X_tune, y_tune,
    categorical_feature=[c for c in categorical_cols_tune if c in X_tune.columns] # Use categorical_cols_tune
)

print("Mejores hiperparámetros encontrados:", grid.best_params_)
print("Mejor F1 Macro (CV):", grid.best_score_)

# -----------------------------
# 4. Entrenamiento final (2010–2021) y test en 2022 con mejores hiperparámetros
# -----------------------------
X_train_val = X[X['censo_anio'] < 2022].drop('censo_anio', axis=1)
y_train_val = y[X_train_val.index]

X_test = X[X['censo_anio'] == 2022].drop('censo_anio', axis=1)
y_test = y[X_test.index]

# Identify categorical columns in X_train_val and X_test
categorical_cols_train_val = X_train_val.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_cols_test = X_test.select_dtypes(include=['object', 'category']).columns.tolist()


lgbm_final = LGBMClassifier(
    n_estimators=1000,
    class_weight="balanced",
    objective="multiclass",
    random_state=42,
    **grid.best_params_
)

# Define callbacks for early stopping and logging for the final model
callbacks_final = [early_stopping(stopping_rounds=50, verbose=-1), log_evaluation(period=100)]

lgbm_final.fit(
    X_train_val, y_train_val,
    categorical_feature=[c for c in categorical_cols_train_val if c in X_train_val.columns], # Use categorical_cols_train_val
    eval_set=[(X_test, y_test)],
    callbacks=callbacks_final # Use callbacks parameter
)

y_pred_test = lgbm_final.predict(X_test)
y_probs_test = lgbm_final.predict_proba(X_test)

print("\n--- Evaluación final en Holdout 2022 ---")
print("Balanced Accuracy:", balanced_accuracy_score(y_test, y_pred_test))
print("F1 Macro:", f1_score(y_test, y_pred_test, average='macro'))

# PR AUC y ROC AUC
classes = lgbm_final.classes_
y_test_bin = label_binarize(y_test, classes=classes)
print("PR AUC (macro):", average_precision_score(y_test_bin, y_probs_test, average="macro"))
print("ROC AUC (macro):", roc_auc_score(y_test_bin, y_probs_test, average="macro"))

print("\nClassification Report:\n", classification_report(y_test, y_pred_test))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_test))

# -----------------------------
# 5. Curvas Precision-Recall en Test 2022
# -----------------------------
n_classes = y_test_bin.shape[1]

plt.figure(figsize=(8, 6))
for i in range(n_classes):
    precision, recall, _ = precision_recall_curve(y_test_bin[:, i], y_probs_test[:, i])
    auc_score = auc(recall, precision)
    plt.plot(recall, precision, label=f'Clase {classes[i]} (AUC = {auc_score:.2f})')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Curvas Precision-Recall en Test 2022 (LightGBM tuned)')
plt.legend()
plt.grid(True)
plt.show()

# -----------------------------
# 6. Evolución de métricas en Rolling Window CV
# -----------------------------
plt.figure(figsize=(8,5))
plt.plot(metrics_df["validation_year"], metrics_df["balanced_accuracy"], marker="o", label="Balanced Accuracy")
plt.plot(metrics_df["validation_year"], metrics_df["f1_macro"], marker="s", label="F1 Macro")
plt.plot(metrics_df["validation_year"], metrics_df["pr_auc"], marker="^", label="PR AUC")
plt.plot(metrics_df["validation_year"], metrics_df["roc_auc"], marker="d", label="ROC AUC")

plt.title("Evolución de métricas en Validación Rolling Window (LightGBM)")
plt.xlabel("Año de validación")
plt.ylabel("Score")
plt.ylim(0,1)
plt.grid(True)
plt.legend()
plt.show()
```

#### Temporal Holdout con Rolling window
Rolling window 2017–2021 para validar estabilidad.

GridSearchCV en 2019–2021 para tuning de hiperparámetros (num_leaves, learning_rate, max_depth, min_child_samples).

Entrenamiento final (2002–2021) y test en 2022 con mejores hiperparámetros

Holdout final 2022 con mejores parámetros.

Métricas: Balanced Accuracy, F1 Macro, PR AUC, ROC AUC.

Curvas PR por clase en el test 2022.

Gráfico de evolución temporal de todas las métricas.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import (
    classification_report, confusion_matrix,
    f1_score, balanced_accuracy_score,
    precision_recall_curve, auc,
    average_precision_score, roc_auc_score
)
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.utils.class_weight import compute_class_weight

# -----------------------------
# 1. Separar features y target
# -----------------------------
X = df_modified.drop(['participo_alteracion_orden_ult_anio_id'], axis=1, errors='ignore')
y = df_modified['participo_alteracion_orden_ult_anio_id']

# Identificar columnas categóricas
categorical_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
for col in categorical_cols:
    X[col] = X[col].astype('str')  # CatBoost requiere strings

# Calculate class weights
classes = np.unique(y)
class_weights = compute_class_weight('balanced', classes=classes, y=y)
class_weights_dict = dict(zip(classes, class_weights))


# -----------------------------
# 2. Rolling Window Validation
# -----------------------------
years = sorted(X['censo_anio'].unique())
print("Años disponibles:", years)

metrics_per_year = []

for val_year in range(2019, 2022):  # validamos de 2019 a 2021
    print(f"\n>>> Entrenando hasta {val_year-1}, validando en {val_year}")

    # Train = todo hasta año previo
    X_train = X[X['censo_anio'] < val_year].drop('censo_anio', axis=1)
    y_train = y[X_train.index]

    # Validation = año actual
    X_val = X[X['censo_anio'] == val_year].drop('censo_anio', axis=1)
    y_val = y[X_val.index]

    # Modelo base
    cat_model = CatBoostClassifier(
        iterations=500,
        learning_rate=0.05,
        depth=8,
        random_seed=42,
        loss_function="MultiClass",
        class_weights=class_weights_dict, # Use the calculated class weights
        verbose=False
    )

    cat_model.fit(
        X_train, y_train,
        cat_features=[X_train.columns.get_loc(c) for c in categorical_cols if c in X_train.columns],
        eval_set=(X_val, y_val),
        early_stopping_rounds=30,
        use_best_model=True
    )

    # Predicciones
    y_val_pred = cat_model.predict(X_val).ravel()
    y_val_proba = cat_model.predict_proba(X_val)

    # Métricas principales
    bal_acc = balanced_accuracy_score(y_val, y_val_pred)
    f1 = f1_score(y_val, y_val_pred, average="macro")

    # PR AUC y ROC AUC (multiclase OVR)
    classes = cat_model.classes_
    y_val_bin = label_binarize(y_val, classes=classes)
    pr_auc = average_precision_score(y_val_bin, y_val_proba, average="macro")
    roc_auc = roc_auc_score(y_val_bin, y_val_proba, average="macro")

    print(f"Balanced Accuracy: {bal_acc:.4f} | F1 Macro: {f1:.4f} | PR AUC: {pr_auc:.4f} | ROC AUC: {roc_auc:.4f}")

    metrics_per_year.append({
        "validation_year": val_year,
        "balanced_accuracy": bal_acc,
        "f1_macro": f1,
        "pr_auc": pr_auc,
        "roc_auc": roc_auc
    })

# Convertir a DataFrame
metrics_df = pd.DataFrame(metrics_per_year)
print("\nResumen Rolling Window CV:")
print(metrics_df)

# -----------------------------
# 3. GridSearchCV en 2020–2021 (para tuning)
# -----------------------------
print("\n>>> Ajustando hiperparámetros con GridSearchCV (2019–2021)")

X_tune = X[X['censo_anio'].isin([2020, 2021])].drop('censo_anio', axis=1)
y_tune = y[X_tune.index]

# Calculate class weights for the tuning data
classes_tune = np.unique(y_tune)
class_weights_tune = compute_class_weight('balanced', classes=classes_tune, y=y_tune)
class_weights_tune_dict = dict(zip(classes_tune, class_weights_tune))


param_grid = {
    "depth": [6, 8, 10],
    "learning_rate": [0.01, 0.05, 0.1],
    "l2_leaf_reg": [1, 3, 5],
    "bagging_temperature": [0.5, 1, 5]
}

cat_grid = CatBoostClassifier(
    iterations=500,
    random_seed=42,
    loss_function="MultiClass",
    class_weights=class_weights_tune_dict, # Use the calculated class weights for tuning
    verbose=False
)

grid = GridSearchCV(
    estimator=cat_grid,
    param_grid=param_grid,
    scoring="f1_macro",
    cv=3,
    n_jobs=-1
)

grid.fit(
    X_tune, y_tune,
    cat_features=[X_tune.columns.get_loc(c) for c in categorical_cols if c in X_tune.columns]
)

print("Mejores hiperparámetros encontrados:", grid.best_params_)
print("Mejor F1 Macro (CV):", grid.best_score_)

# -----------------------------
# 4. Entrenamiento final (2010–2021) y test en 2022 con mejores hiperparámetros
# -----------------------------
X_train_val = X[X['censo_anio'] < 2022].drop('censo_anio', axis=1)
y_train_val = y[X_train_val.index]

X_test = X[X['censo_anio'] == 2022].drop('censo_anio', axis=1)
y_test = y[X_test.index]

# Calculate class weights for the final training data
classes_train_val = np.unique(y_train_val)
class_weights_train_val = compute_class_weight('balanced', classes=classes_train_val, y=y_train_val)
class_weights_train_val_dict = dict(zip(classes_train_val, class_weights_train_val))


cat_final = CatBoostClassifier(
    iterations=1000,
    random_seed=42,
    loss_function="MultiClass",
    class_weights=class_weights_train_val_dict, # Use the calculated class weights for the final model
    verbose=False,
    **grid.best_params_
)

cat_final.fit(
    X_train_val, y_train_val,
    cat_features=[X_train_val.columns.get_loc(c) for c in categorical_cols if c in X_train_val.columns],
    eval_set=(X_test, y_test),
    early_stopping_rounds=50,
    use_best_model=True
)

y_pred_test = cat_final.predict(X_test).ravel()
y_probs_test = cat_final.predict_proba(X_test)

print("\n--- Evaluación final en Holdout 2022 ---")
print("Balanced Accuracy:", balanced_accuracy_score(y_test, y_pred_test))
print("F1 Macro:", f1_score(y_test, y_pred_test, average='macro'))

# PR AUC y ROC AUC
classes = cat_final.classes_
y_test_bin = label_binarize(y_test, classes=classes)
print("PR AUC (macro):", average_precision_score(y_test_bin, y_probs_test, average="macro"))
print("ROC AUC (macro):", roc_auc_score(y_test_bin, y_probs_test, average="macro"))

print("\nClassification Report:\n", classification_report(y_test, y_pred_test))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_test))

# -----------------------------
# 5. Curvas Precision-Recall en Test 2022
# -----------------------------
n_classes = y_test_bin.shape[1]

plt.figure(figsize=(8, 6))
for i in range(n_classes):
    precision, recall, _ = precision_recall_curve(y_test_bin[:, i], y_probs_test[:, i])
    auc_score = auc(recall, precision)
    plt.plot(recall, precision, label=f'Clase {classes[i]} (AUC = {auc_score:.2f})')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Curvas Precision-Recall en Test 2022 (CatBoost tuned)')
plt.legend()
plt.grid(True)
plt.show()

# -----------------------------
# 6. Evolución de métricas en Rolling Window CV
# -----------------------------
plt.figure(figsize=(8,5))
plt.plot(metrics_df["validation_year"], metrics_df["balanced_accuracy"], marker="o", label="Balanced Accuracy")
plt.plot(metrics_df["validation_year"], metrics_df["f1_macro"], marker="s", label="F1 Macro")
plt.plot(metrics_df["validation_year"], metrics_df["pr_auc"], marker="^", label="PR AUC")
plt.plot(metrics_df["validation_year"], metrics_df["roc_auc"], marker="d", label="ROC AUC")

plt.title("Evolución de métricas en Validación Rolling Window (CatBoost)")
plt.xlabel("Año de validación")
plt.ylabel("Score")
plt.ylim(0,1)
plt.grid(True)
plt.legend()
plt.show()
```
